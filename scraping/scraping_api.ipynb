```python
import requests

cookies = {
    '_ga': 'GA1.1.1408021629.1724975010',
    'auth-production': 's%3A2mScyHJKumX8WD9eNMNoIbkI1FdgdNtw.c9652l9mbgkQTYhQaoCGHN4XFFSib6aZG27Zdq7RkeE',
}

headers = {
    'accept': 'application/json, text/plain, */*',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36',
    'x-requested-with': 'XMLHttpRequest',
}

all_results = []
skip = 0
limit = 40

while True:
    url = f'https://ruta100.cl/car/filter?search=&locations=&brand=&model=&type=&transmission=&fuelType=&doors=&year=,&kms=,&price=,&seals=&region=&commune=&sort=&new=&skip={skip}&limit={limit}'
    
    response = requests.get(url, cookies=cookies, headers=headers)
    data = response.json()
    
    results = data.get('payload', {}).get('results', [])
    
    if not results:
        break  # Sale del bucle si no hay m√°s resultados
    
    all_results.extend(results)
    skip += limit  # Incrementa para la siguiente p√°gina

print(f"Total de autos obtenidos: {len(all_results)}")
```

```python
import pandas as pd

# Inicializamos listas vac√≠as para almacenar los datos
model = []
price = []
kilometraje = []
city = []
cylinder = []
region = []

# Iteramos sobre todos los resultados obtenidos de varias p√°ginas
for result in all_results:  # Usamos 'all_results' que contiene todos los autos
    attributes = result['attributes']
    model.append(attributes.get('model'))
    price.append(attributes.get('price'))
    kilometraje.append(attributes.get('kms'))
    city.append(attributes.get('commune'))
    cylinder.append(attributes.get('engineSize', ''))
    region.append(attributes.get('region'))

# Creaci√≥n del DataFrame con todos los autos obtenidos
df = pd.DataFrame({
    'model': model,
    'price': price,
    'kilometraje': kilometraje,
    'cylinder': cylinder,
    'city': city,
    'region': region
})

# Mostrar las primeras filas del DataFrame
df.head()
```

```python
import pandas as pd

# Inicializamos listas vac√≠as para almacenar los datos
model = []
price = []
kilometraje = []
city = []
cylinder = []
region = []
year = []
fuel_type = []
transmission = []
doors = []
plate = []
brand = []
vehicle_type = []
description = []
version = []
only_owner = []
never_bumped = []
checked_maintenance = []
state = []

# Iteramos sobre todos los resultados obtenidos de varias p√°ginas
for result in all_results:
    attributes = result['attributes']
    model.append(attributes.get('model'))
    price.append(attributes.get('price'))
    kilometraje.append(attributes.get('kms'))
    city.append(attributes.get('commune'))
    cylinder.append(attributes.get('engineSize', ''))
    region.append(attributes.get('region'))
    year.append(attributes.get('year'))
    fuel_type.append(attributes.get('fuelType', ''))
    transmission.append(attributes.get('transmission', ''))
    doors.append(attributes.get('doors', ''))
    plate.append(attributes.get('plate', ''))
    brand.append(attributes.get('brand', ''))
    vehicle_type.append(attributes.get('type', ''))
    description.append(result.get('description', ''))
    version.append(attributes.get('version', ''))
    only_owner.append(result.get('onlyOwner', ''))
    never_bumped.append(result.get('neverBumped', ''))
    checked_maintenance.append(result.get('checkedMaintenance', ''))
    state.append(result.get('state', ''))

# Creaci√≥n del DataFrame con todos los atributos
df = pd.DataFrame({
    'brand': brand,
    'model': model,
    'version': version,
    'price': price,
    'kilometraje': kilometraje,
    'cylinder': cylinder,
    'city': city,
    'region': region,
    'year': year,
    'fuel_type': fuel_type,
    'transmission': transmission,
    'doors': doors,
    'plate': plate,
    'description': description,
    'only_owner': only_owner,
    'never_bumped': never_bumped,
    'checked_maintenance': checked_maintenance,
    'state': state
})

df.head()
```
